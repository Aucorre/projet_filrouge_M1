```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('merged_data.csv')

df
```

```python
df["customer_id"].nunique()
```

```python
single_order_clients = df.groupby('customer_unique_id').filter(lambda x: len(x) == 1)
single_order_clients
```

```python
import datetime as dt

df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])


df["order_approved_at"] = pd.to_datetime(df["order_approved_at"])
current_date = max(df["order_approved_at"])

rfm = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (current_date - x.max()).days,
    'order_id': 'count',
    'payment_value': 'sum'
}).reset_index()


rfm.columns = ['customer_unique_id', 'Recency', 'Frequency', 'Monetary']

rfm.head()
```

```python
rfm.sort_values("Monetary")
```

```python
rfm["RecencyScore"] = pd.qcut(rfm['Recency'], 4, ['1', '2', '3', '4'])
rfm["FrequencyScore"] = pd.qcut(rfm['Frequency'].rank(method="first"), 4, ['4', '3', '2', '1'])
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 4, ['4', '3', '2', '1'])



rfm["RFM"] = rfm["RecencyScore"].astype(str) + rfm["FrequencyScore"].astype(str) + rfm["MonetaryScore"].astype(str)

rfm["RFM_SCORE"] = rfm["RecencyScore"].astype(int) + rfm["FrequencyScore"].astype(int) + rfm["MonetaryScore"].astype(int)

rfm["Segment"] = pd.qcut(
    rfm["RFM_SCORE"], 
    q=[0, 0.04, 0.25, 0.5, 0.75, 1],
    labels=["VIP", "TBC", "BC", "CC", "Other"],
    duplicates="drop"
)

segment_distribution = rfm["Segment"].value_counts()
print(segment_distribution)
```

```python
df["Customer Segment"] = df["customer_unique_id"].map(rfm.set_index("customer_unique_id")["Segment"])
```

```python
df.head()
```

```python
df['customer_mean_review_score'] = df['customer_unique_id'].map(df.groupby('customer_unique_id')['review_score'].mean())
```

```python
df["Unhappy"] = (df["Customer Segment"] == "VIP") & (df["order_purchase_timestamp"] > current_date - dt.timedelta(days=30)) & (df["review_score"] < df ["customer_mean_review_score"])

df.head()
```
